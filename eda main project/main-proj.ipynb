```python
df[df['Release Date'] == 'releases on TBD'] = 'Dec 30, 2024'
```

```python
df['Release Date'] = pd.to_datetime(df['Release Date'])
df['month'] = df['Release Date'].dt.month
df['year'] = df['Release Date'].dt.year
df['day'] = df['Release Date'].dt.day
```

```python
plt.figure(figsize=(16, 10))
plt.title('Number of released games with respect to month')
sns.countplot(data=df, y='month')
plt.show()
```

```python
plt.figure(figsize=(16, 10))
plt.title('Number of released games with respect to month')
sns.countplot(data=df, y='year')
plt.show()
```

```python
columns_to_change = ['Times Listed', 'Number of Reviews', 'Plays',
                     'Playing', 'Backlogs', 'Wishlist']
```

```python
for col in columns_to_change:
    tmp_col = []
    for val in df[col]:
        if isinstance(val, str):
            if 'K' in val:
                tmp_col.append(float(val.replace('K', '')) * 1000)
            else:
                tmp_col.append(float(val))
        else:
            tmp_col.append(val)
    df[col] = tmp_col

```

```python
df.head(2)
```

```python
# Set the plot size
plt.figure(figsize=(16, 10))

# Loop through the columns to create a subplot for each column
for count, col in enumerate(columns_to_change):
    # Create a subplot
    plt.subplot(3, 2, count + 1)
    
    # Sort the values in the current column and get the index of the top 10 values
    indexes = df[col].sort_values(ascending=False)[:10].index
    
    # Get a new DataFrame with the top 10 games and their values in the current column
    tmp_df = df.loc[indexes][['Title', col]]
    
    # Create a bar plot with the values in the current column on the x-axis and the game titles on the y-axis
    sns.barplot(data=tmp_df, x=col, y='Title')
    
    # Set the title for the current subplot
    plt.title(f'Top 10 games according to "{col}"')

# Adjust the layout of the subplots to fit nicely on the figure
plt.tight_layout()

# Show the plot
plt.show()
```

```python
genres_dict = {}
for index, row in df.iterrows():
    x = str(row['Genres']).replace('[', '').replace(']', '').replace("'", '').split(',')
    
    for genre in x:
        if genre not in genres_dict:
            genres_dict[genre] = 1
        else:
            genres_dict[genre] += 1
```

```python
tmp_df = pd.DataFrame({'genre': genres_dict.keys(),
              'count': genres_dict.values()})
```

```python
plt.figure(figsize=(16,10))
sns.barplot(data= tmp_df, y='genre', x='count')
```
