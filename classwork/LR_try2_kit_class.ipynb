```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel('LR_try2_kit_class.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
plt.scatter(df['x'] , df['y'])
plt.xlabel('x')
plt.ylabel("y")
plt.title("x vs y")
```

```python
X = df.iloc[0:5 , 0].values.reshape(-1,1)
Y = df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression().fit(X,Y)
```

```python
print("Intercept:" , lr.intercept_)
print("Slope:",lr.coef_)
```

```python
reg_line=lr.coef_*X+lr.intercept_
plt.figure(figsize=(12,10))
plt.title("X vs Y",fontsize=20,color="Blue")
plt.scatter(X,Y,label="Model",color='red')
plt.plot(X,reg_line,label="Linear Regression",color="green")
```
